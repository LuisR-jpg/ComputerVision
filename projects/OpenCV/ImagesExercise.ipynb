# HW regarding images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

1. Crea un script que lea 5 imagenes

```python
nImages = 5
images = []
for i in range(1, nImages + 1):
    images.append(plt.imread(f'images/imagen_{i}.jpg'))
    plt.subplot(nImages // 2, 2, i)
    plt.imshow(images[i - 1])
```

2. Transformar a gris y que se muestren

```python
for i, img in enumerate(images):
    plt.subplot((nImages + 1) // 2, 2, i + 1)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2GRAY), cmap = "gray")
```

3. Tomar la primer imagen y aplicar Canny con `[100, 200, 300, 400, 500]`

```python
values = [100, 200, 300, 400, 500]
plt.figure(figsize=(15, 15))
for i, v in enumerate(values):
    plt.subplot((len(values) + 1) // 2, 2, i + 1)
    plt.imshow(cv.Canny(images[1], v, v), cmap = 'gray')
    
```

4. Resize height a 300 px sin deformar la imagen

```python
plt.figure(figsize=(15, 15))
for i, img in enumerate(images):
    fy = 300 / img.shape[0]
    plt.subplot((len(images) + 1) // 2, 2, i + 1)
    plt.imshow(cv.resize(img, (0, 0), fx = fy, fy = fy))
```
